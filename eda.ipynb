```python
# Set seaborn style
sns.set(style="whitegrid")

# 1. Load Data
file_path = "StudentsPerformance.csv"  # Change if needed
df = pd.read_csv(file_path)
```

```python
# 2. Overview of Data
print("===== Head of Dataset =====")
print(df.head())

print("\n===== Info =====")
print(df.info())

print("\n===== Describe (Numerical) =====")
print(df.describe())
```

```python
# 3. Check for Missing Values
print("\n===== Missing Values =====")
print(df.isnull().sum())
```

```python
# 4. Categorical Variables Distribution
categorical_columns = ["gender", "race/ethnicity", "parental level of education", 
                       "lunch", "test preparation course"]

for col in categorical_columns:
    print(f"\n===== Value Counts for {col} =====")
    print(df[col].value_counts())
    sns.countplot(x=col, data=df)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

```python
# 5. Grouped Statistics (mean math score by group)
groups = ["gender", "race/ethnicity", "parental level of education", "test preparation course"]

for group in groups:
    print(f"\n===== Mean Scores by {group} =====")
    print(df.groupby(group)[["math score", "reading score", "writing score"]].mean())
```

```python
# 6. Boxplots of scores by gender
plt.figure(figsize=(12, 4))
for i, subject in enumerate(["math score", "reading score", "writing score"]):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x="gender", y=subject, data=df)
    plt.title(f"{subject} by Gender")
plt.tight_layout()
plt.show()
```

```python
# 7. Histograms for score distributions
plt.figure(figsize=(12, 4))
for i, subject in enumerate(["math score", "reading score", "writing score"]):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[subject], bins=15, kde=True)
    plt.title(f"Distribution of {subject}")
plt.tight_layout()
plt.show()
```

```python
# 8. Correlation Heatmap
plt.figure(figsize=(6, 5))
corr = df[["math score", "reading score", "writing score"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=0.5, vmax=1.0)
plt.title("Correlation Matrix of Scores")
plt.show()
```
